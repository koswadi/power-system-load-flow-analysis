# Newton-Raphson Load Flow Analysis

This notebook demonstrates the Newton-Raphson method for solving power flow equations.

```python
import sys
sys.path.append("../src")

from ybus_builder import YBusBuilder
from newton_raphson import NewtonRaphsonSolver
```

```python
ybus = YBusBuilder(
    "../data/sample_line_data.csv"
).build()

solver = NewtonRaphsonSolver(ybus)

voltages, angles, iterations = solver.solve()
```

```python
print("Iterations:", iterations)

for i in range(len(voltages)):
    print(
        f"Bus {i+1}: "
        f"Voltage={voltages[i]:.4f}, "
        f"Angle={angles[i]:.4f}"
    )
```

## Discussion

Newton-Raphson generally converges faster than Gauss-Seidel and is the preferred method for large-scale power systems.
