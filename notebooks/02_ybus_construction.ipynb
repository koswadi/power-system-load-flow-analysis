# Y-Bus Matrix Construction

This notebook builds the bus admittance matrix (Y-Bus).

Y-Bus is the foundation of load flow analysis.

```python
builder = YBusBuilder("../data/sample_line_data.csv")

ybus = builder.build()

print(ybus)
```

## Interpretation

Diagonal elements represent self-admittance.

Off-diagonal elements represent mutual coupling between buses.
