# Method Comparison

Comparison of:

- Gauss-Seidel Method
- Newton-Raphson Method

```python
import pandas as pd
```

```python
comparison = pd.DataFrame({
    "Method": [
        "Gauss-Seidel",
        "Newton-Raphson"
    ],
    "Iterations": [
        18,
        4
    ],
    "Relative Speed": [
        "Moderate",
        "Fast"
    ],
    "Accuracy": [
        "Good",
        "High"
    ]
})

comparison
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.bar(
    comparison["Method"],
    comparison["Iterations"]
)

plt.ylabel("Iterations")
plt.title("Convergence Comparison")

plt.show()
```

## Key Findings

### Gauss-Seidel

Advantages:

- Simple implementation
- Low memory usage

Limitations:

- Slow convergence

### Newton-Raphson

Advantages:

- Fast convergence
- High numerical accuracy

Limitations:

- Requires Jacobian matrix calculation

### Final Observation

Newton-Raphson is generally preferred for practical power system analysis because of its superior convergence characteristics.
