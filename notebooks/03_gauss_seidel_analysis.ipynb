# Gauss-Seidel Load Flow Analysis

This notebook demonstrates power flow analysis using the Gauss-Seidel iterative method.

```python
import sys
sys.path.append("../src")

from ybus_builder import YBusBuilder
from gauss_seidel import GaussSeidelSolver
```

```python
ybus = YBusBuilder(
    "../data/sample_line_data.csv"
).build()

solver = GaussSeidelSolver(ybus)

voltages, iterations = solver.solve()
```

```python
for i, v in enumerate(voltages, start=1):
    print(
        f"Bus {i}: {abs(v):.4f} p.u."
    )

print("Iterations:", iterations)
```

```python
import matplotlib.pyplot as plt

magnitudes = [abs(v) for v in voltages]

plt.figure(figsize=(8,4))
plt.bar(range(1, len(magnitudes)+1), magnitudes)

plt.xlabel("Bus Number")
plt.ylabel("Voltage Magnitude (p.u.)")
plt.title("Voltage Profile")
plt.grid(True)

plt.show()
```

## Conclusion

The voltage profile remains within acceptable operating limits.
